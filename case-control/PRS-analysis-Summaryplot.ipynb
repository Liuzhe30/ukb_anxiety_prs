## Summary and plot

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import json
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
# load feature importance file
with open("datafile/top20-feature-DT-gut.json") as f:
    DT_dict_gut = json.load(f)
with open("datafile/top20-feature-DT-pathway.json") as f:
    DT_dict_pathway = json.load(f)
with open("datafile/top20-feature-RF-gut.json") as f:
    RF_dict_gut = json.load(f)
with open("datafile/top20-feature-RF-pathway.json") as f:
    RF_dict_pathway = json.load(f)
with open("datafile/top20-feature-XGB-gut.json") as f:
    XGB_dict_gut = json.load(f)
with open("datafile/top20-feature-XGB-pathway.json") as f:
    XGB_dict_pathway = json.load(f)
with open("datafile/top20-feature-lightGBM-gut.json") as f:
    lightGBM_dict_gut = json.load(f)
with open("datafile/top20-feature-lightGBM-pathway.json") as f:
    lightGBM_dict_pathway = json.load(f)
```

### respective visulization

```python
# DT gut
name_list = []
pval_list = []
imp_list = []
for item in DT_dict_gut:
    name_list.append("GCST90027" + DT_dict_gut[item]['index'] + ": " + DT_dict_gut[item]['name'].split('.')[-1])
    pval_list.append(float(DT_dict_gut[item]['pvalue(-logP)']))
    imp_list.append(DT_dict_gut[item]['feature importance']*10000)

bar_width = 0.45
x = np.arange(20)[::-1]
pal = sns.color_palette("Purples",40)
pal2 = sns.color_palette("Greens_d",20)
rank = np.arange(20)[::-1]
print(rank)

#ax = sns.barplot(y=x, x=gut_imp_list, label="feature importance (1000)", palette=np.array(pal)[rank], orient='h')
#ax = sns.barplot(y=x+bar_width, x=gut_pval_list, label="pvalue(-logP)", palette=np.array(pal2)[rank], orient='h')
plt.barh(x, pval_list, bar_width, align="center", label="pvalue(-logP)", color='#EDB120')
plt.barh(x+bar_width, imp_list, bar_width, align="center", label="feature importance (10000)", color=np.array(pal)[rank+20])

plt.xlabel("values (Decision Tree)")
plt.ylabel("gut microbiota")
plt.legend(loc = 4)
plt.yticks(x + bar_width / 2, name_list)
plt.savefig("figures/DT-gut.png",dpi=300,bbox_inches = 'tight')
plt.show()
```
